```python
#데이터 구성 패키지 : Series, DataFrame
import datetime 
import pandas as pd
#행렬 연산 패키지
import numpy as np

#경고메세지 무시
import warnings
warnings.filterwarnings(action='ignore')


%matplotlib inline
```

```python
df_sales=pd.read_csv('./sales_member_join_processed.csv')
```

```python
df_sales.head()
```

```python
#데이터 올라옴!
df_sales['purchase_date'] = pd.to_datetime(df_sales['purchase_date'])
```

```python
def how_much(x):
    q1,q2,q3=3.732500*1e04,1.419500*1e05,5.332500*1e05
    if x<q1:
        group = 'low'
    elif q1<=x<q2:
        group = 'middle'
    elif q2<=x<q3:
        group = 'high'
    else:
        group = 'super_high'
        
    return group


df_2= df_sales.groupby('customer_id').sum()['purchase_amt'].sort_values()
df_2=df_2.reset_index()
df_2['spent_group'] = list(map(how_much,df_2['purchase_amt']))
```

```python
df_superhigh=df_2.loc[(df_2['spent_group']=='super_high')|(df_2['spent_group']=='high')]
```

```python
df_superhigh = df_superhigh.drop(['purchase_amt','purchase_amt'],axis=1)
```

```python
df_sum=pd.merge(df_sales,df_superhigh,left_on='customer_id',right_on='customer_id',how='inner')
df_sum.shape
```

```python
df_sum=df_sum.loc[df_sum['purchase_date']>='2019-01-01']
```

```python
df_sum
```

```python
df_final=df_sum.groupby('customer_id').agg('sum')
```

```python
df_final = df_final.reset_index()
df_final.to_csv('superhigh')
```
